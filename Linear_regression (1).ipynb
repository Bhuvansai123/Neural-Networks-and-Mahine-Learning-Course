```python
import pandas as pd
df=pd.read_csv("towns.csv")
df
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
### first we imported LabelEncoder and declared object
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
## fit_trandform() function will categorise the dfle.town as it is the column we have to apply dummies
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```

```python
##.values will give the array
X=dfle[['town','area']].values
X
```

```python
y=dfle.price.values
```

### Now use one hot encoder to create dummy variables for each of the town

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
### we are giving the [0] to say 0th column is my categorical column in X
## refer book for syntax
ct=ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')
ct
```

```python
X=ct.fit_transform(X)
```

```python
## we have to delete one dummy column so X[:,1:] removes 0 column which is dummy column
X=X[:,1:]
X
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.25,random_state=40)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict([[0,1,3400]])  # 3400 sqr ft home in west windsor
```

```python
model.predict([[0,1,3400]])  # 2800 sqr ft home in robbinsville
```

```python
model.score(xtest,ytest)
```
